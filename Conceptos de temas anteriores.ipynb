<h1><b style="font-size: 18px;">FUNDAMENTOS DEL MOVIMIENTO DE FLUIDOS</b></h1>
<br>
<hr>  
    
<p style="font-size: 15px;text-align: justify;">Flujo: Movimiento de un fluido.</p>
    
<p style="font-size: 15px;text-align: justify;">En función del intercambio de la cantidad de movimiento entre moléculas:</p>
    
<p style="font-size: 15px;text-align: justify;">
-Flujo laminar: Las trayectorias del flujo se mueven a lo largo de trayectorias lisas en capas o láminas, deslizándose una capa sobre la adyacente, sin intercambio de cantidad de movimiento y cumpliéndose la ley de Newton de la viscosidad. <br> 
-Flujo turbulento: Las partículas de fluido se mueven siguiendo trayectorias irregulares, originando un intercambio de cantidad de movimiento de una porción de fluido a otra.</p>

<br>
    
<p><b style="font-size: 14px;">FLUJO VOLUMÉTRICO Y FLUJO MÁSICO</b></p>
    
<p style="font-size: 15px;text-align: justify;">
-Flujo volumétrico: Volumen de fluido que atraviesa una sección en la unidad de tiempo.  </p>
    
<p style="font-size: 15px;text-align: justify;">
\begin{equation}
    Q =  A \cdot v \hspace{3mm} \left[\frac{m^{^3}}{s}\right]
\end{equation}  </p>  
    
<p style="font-size: 15px;text-align: justify;">
-Flujo másico: Masa de fluido que atraviesa una sección por unidad de tiempo.  </p> 
    
<p style="font-size: 15px;text-align: justify;">
\begin{equation}
    \dot{m} = \rho \cdot v \cdot A = \rho \cdot Q \hspace{3mm} \left[\frac{kg}{s}\right]
\end{equation} </p> 
    
<h1><b style="font-size: 18px;">ECUACIÓN DE BERNOULLI (Tema 12)</b></h1>
<br>
<hr>  
    
<p style="font-size: 15px;text-align: justify;">La ecuación de Bernoulli se aplica a flujos incompresibles y permanentes, sin fricción.</p>
    
<br>
    
<p style="font-size: 15px;text-align: justify;">
\begin{equation}
    k' = z + \frac{P}{\rho \cdot g} + \frac{v^{^2}}{2 \cdot g} \hspace{10mm} \left[\frac{Energía}{Peso}\right] = \frac{J}{N} = \frac{N \cdot m}{N} = mcl
\end{equation} </p>
