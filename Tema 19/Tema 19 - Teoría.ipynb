<h1><b style="font-size: 18px;">PÉRDIDAS MENORES</b></h1>
<br>
<hr>  
    
<p style="font-size: 15px;text-align: justify;">Las pérdidas menores son pérdidas de carga generadas en válvulas, codos, salida de depósitos, entrada de depósitos, estrechamientos, es decir, en las piezas que no son tubería y se encuentran en la instalación. </p>
    
<p style="font-size: 15px;text-align: justify;">Para calcular las pérdidas de carga hay dos métodos: </p>

<br>
    
<p><b style="font-size: 14px;">LONGITUD EQUIVALENTE</b></p>
    
<p style="font-size: 15px;text-align: justify;">Longitud de tubería que, con el mismo diámetro que el de la pieza, produce igual pérdida de carga. </p>
    
<img src="../img/longitud_equivalente.png" style="height:80px">

<p style="font-size: 15px;text-align: justify;">Pérdidas de carga mediante la expresión de Darcy-Weisbach, en el caso de piezas especiales: </p>
    
<p style="font-size: 15px;text-align: justify;">
\begin{equation}
    h_{_{f}} = f \cdot \frac{L_{eq}}{D} \cdot \frac{{v}^{2}}{2 \cdot g} 
\end{equation} </p>
    
<p style="font-size: 15px;text-align: justify;">En el ábaco nº23 se obtienen las longitudes equivalentes de las piezas especiales según el diámetro. </p>
    
<br>
    
<p><b style="font-size: 14px;">FACTOR DE PASO</b></p>

<p style="font-size: 15px;text-align: justify;">
\begin{equation}
    h_{_{f}} = k \cdot \frac{{v}^{2}}{2 \cdot g} 
\end{equation} </p>    

<p style="font-size: 15px;text-align: justify;">Siendo k el factor de paso adimensional en piezas especiales. </p>
    
<p style="font-size: 15px;text-align: justify;">En el cuadro nº 24 se encuentran los factores de paso para diferentes piezas especiales. </p>
    
<br>
    
<h1><b style="font-size: 18px;">ENVEJECIMIENTO DE TUBERÍAS</b></h1>
<br>
<hr>     
 
<p style="font-size: 15px;text-align: justify;">El tiempo de funcionamiento influye en la rugosidad absoluta ε del material, en el coeficiente de frotamiento f y, por tanto, en las pérdidas de carga $h_{f}$. </p> 
    
<p style="font-size: 15px;text-align: justify;">
\begin{equation}
    h_{_{f}}(t) = h_{_{f}}(0) \cdot \left[ 1+ \frac{t}{100} \right] 
\end{equation} </p>    
    
<p style="font-size: 15px;text-align: justify;">t= tiempo en años</p> 
    
<p style="font-size: 15px;text-align: justify;">Se estima que las pérdidas de carga generadas por una tubería aumentan un 1% cada año transcurrido. </p>     
    
<br>
    
<h1><b style="font-size: 18px;">LÍNEA PIEZOMÉTRICA Y DE ALTURA TOTAL</b></h1>
<br>
<hr>     
    
<p><b style="font-size: 14px;">ALTURA PIEZOMÉTRICA</b></p>
    
<p style="font-size: 15px;text-align: justify;">Suma de la energía de cota y la energía de presión. </p>     
    
<p style="font-size: 15px;text-align: justify;">Altura piezométrica: $ z + \frac{P}{\gamma}$ </p> 
    
<p><b style="font-size: 14px;">ALTURA TOTAL</b></p>
    
<p style="font-size: 15px;text-align: justify;">Resultante de sumar a la altura piezométrica la energía de velocidad.</p>     
    
<p style="font-size: 15px;text-align: justify;">Altura total: $z + \frac{P}{\gamma} + \frac{v^{^{2}}}{2 \cdot g}$ </p> 
    
<br>
    
<h1><b style="font-size: 18px;">FÓRMULA DE HAZEN-WILLIAMS</b></h1>
<br>
<hr>     
     
<p style="font-size: 15px;text-align: justify;">
\begin{equation}
    h_{f} = J_{_{1}} \cdot L \cdot {Q_{(l/s)}}^{^{1.852}}
\end{equation} </p>
    
<p style="font-size: 15px;text-align: justify;">Válida cuando $\upsilon$ del fluido es cercana a $1 \cdot 10^{-6}$ $\frac{m^{2}}{s}$ </p> 
    
<p style="font-size: 15px;text-align: justify;">La tabla de las pérdidas de carga unitarias $(J_{1})$ se encuentra en el cuadro nº25, y depende del diámetro de la tubería y del coeficiente de Hazen-Williams $(C_{HW})$ </p>  
    
<p style="font-size: 15px;text-align: justify;">
\begin{equation}
    J_{_{1}} = \frac{1.2117 \cdot 10^{^{10}}}{{C_{HW}}^{1.852} \cdot {D_{mm}}^{4.87}}
\end{equation} </p>     
    
<br>
    
<p style="font-size: 15px;text-align: justify;">El coeficiente de Hazen-Williams se encuentra en el cuadro 25. </p>
