<h1><b style="font-size: 18px;">ECUACIÓN DE DARCY-WEISBACH</b></h1>
<br>
<hr>  
    
<p style="font-size: 15px;text-align: justify;">Deducción de la ecuación de Darcy Weisbach, utilizada para calcular pérdidas de carga en tuberías.</p>

<p style="font-size: 15px;text-align: justify;">
Hipótesis de partida: 
<br><br>
- Flujo turbulento<br>
- Régimen permanente y uniforme <br>
- Fluido incompresible<br>
- Conducto cerrado<br>
- Flujo unidimensional<br>
<br></p>
    
<p style="font-size: 8px;text-align: justify;"><b style="font-size: 18px;">- Teorema de la cantidad de movimiento</b></p> 
    
<p style="font-size: 15px;text-align: justify;">Se selecciona un volumen de control y se estudian las fuerzas que actuan sobre dicho volumen.</p>

<img src="../img/Darcy-Weisbach.png" style="height:200px">
    
<p style="font-size: 15px;text-align: justify;">
1. Vc <br>
2. Ejes <br>
3. Fuerzas exteriores $\left\{ \begin{array}{c} W \\P_{_{1}} \cdot A,~ P_{_{2}} \cdot A \\ R_{x}, R_{y}  \end{array}\right. $ <br><br>
$\tau = \frac{F}{A} = \frac{R_{x}}{P \cdot L} $ por lo tanto, $ R_{x} = \tau \cdot P \cdot L $ <br><br><br>
4. Velocidades
<br><br>
$\left. \begin{array}{c} A = A_{_{1}} = A_{_{2}} = cte\\ Q = cte \end{array}\right\} $ $v = v_{_{1}} = v_{_{2}} = cte$ <br><br>
5. <b style="font-size: 16px;">Eje x</b></p>

<p style="font-size: 15px;text-align: justify;">
    
\begin{align}
    \sum F_{x} &= 0 \\
    P_{_{1}} \cdot A_{_{1}} - P_{_{2}} \cdot A_{_{2}} + \underbrace{W}_{\gamma \cdot \underbrace{V}_{A \cdot L}} \cdot \sin(\theta) - \tau \cdot P \cdot L &= \rho \cdot \left[Q \cdot v - Q \cdot v\right]\\
    \left(P_{_{1}} - P_{_{2}} \right) \cdot A + \gamma \cdot A \cdot \underbrace{L \cdot \sin(\theta)}_{z_{_{1}} - z_{_{2}}} &= \tau \cdot P \cdot L \\    
    Dividir ~ por \left(\gamma \cdot A \right): \frac{P_{_{1}} - P_{_{2}}}{\gamma} + \left(z_{_{1}} - z_{_{2}} \right) &= \frac{\tau \cdot P \cdot L}{\gamma \cdot A} \hspace{1.5cm} (1)
\end{align}</p>
    
<p style="font-size: 8px;text-align: justify;"><b style="font-size: 18px;">- Bernoulli</b></p> 

<p style="font-size: 15px;text-align: justify;">
\begin{align}
    B_{_{1}} - h_{f_{12}} &= B_{_{2}}\\
    z_{_{1}} + \frac{P_{_{1}}}{\gamma} + \frac{{v_{_{1}}}^{^2}}{2 \cdot g} - h_{f_{12}} &= z_{_{_2}} + \frac{P_{_{2}}}{\gamma} + \frac{{v_{_{2}}}^{^2}}{2 \cdot g}\\
    \frac{P_{_{1}} - P_{_{2}}}{\gamma} + \left(z_{_{1}} - z_{_{2}} \right) &= h_{f_{12}}\hspace{5.6cm} (2)
\end{align}</p>

<br>

<p style="font-size: 15px;text-align: justify;">
\begin{equation}
    \left(1 \right) ~ y ~ \left(2 \right) \rightarrow h_{f_{12}} = \frac{\tau \cdot P \cdot L}{\gamma \cdot A}
\end{equation}</p>
    
<br>
    
<p style="font-size: 15px;text-align: justify;">
Según la <b>teoría de la longitud de mezcla Prandtl:</b> $ \tau = \lambda \cdot \rho \cdot \frac{v^2}{2}$
    
<br>

($ \lambda $ es un factor de proporcionalidad adimensional)</p>
    
<br>
    
<p style="font-size: 15px;text-align: justify;">
\begin{equation}
    h_{f_{12}} = \left(\lambda \cdot \rho \cdot \frac{v^{^{2}}}{2} \right)\cdot \frac{P \cdot L}{\gamma \cdot A} = \lambda \cdot \frac{v^{^{2}}}{2 \cdot g} \cdot \frac{L}{\frac{A}{P}} 
\end{equation}</p>
    
<br>
    
<p style="font-size: 15px;text-align: justify;">
Siendo el <b>radio hidráulico $R_h$</b> cociente entre el área de la sección transversal del flujo (A) y el perímetro mojado (P), $R_h = \frac{A}{P}$, se tiene: </p>
    
<br>
    
<p style="font-size: 15px;text-align: justify;">
$$ \boxed{h_{f_{12}} = \lambda \cdot \frac{v^{^{2}}}{2 \cdot g} \cdot \frac{L}{R_{h}}}$$</p>
    
<br>
    
<p style="font-size: 15px;text-align: justify;">
Expresión de Darcy-Weisbach, para el caso de conductos cerrados de sección cualquiera.  
  
Para el caso de tubos circulares el valor del radio hidráulico es: $R_h = \frac{A}{P} = \frac{\frac{\pi \cdot D^{^{2}}}{4}}{\pi \cdot D} = \frac{D}{4}$ </p>

<p style="font-size: 15px;text-align: justify;">
$$ \boxed{h_{f_{12}} = 4 \cdot \lambda \cdot \frac{v^{^{2}}}{2 \cdot g} \cdot \frac{L}{D} = f \cdot \frac{v^{^{2}}}{2 \cdot g} \cdot \frac{L}{D}}$$</p>
    
<p style="font-size: 15px;text-align: justify;">  
Expresión de Darcy-Weisbach para <b>conductos circulares.</b></p>

<h1><b style="font-size: 18px;">TUBOS LISOS Y RUGOSOS DESDE EL PUNTO DE VISTA HIDRÁULICO. FRONTERAS</b></h1>
<br>
<hr> 

<p style="font-size: 15px;text-align: justify;">
El espesor de la subcapa laminar es inversamente proporcional al número de Reynolds. Al aumentar Re disminuye la subcapa laminar $\delta_{sl}$.</p>

<p style="font-size: 15px;text-align: justify;">
Los valores de espesor medio de la rugosidad de las paredes del conducto ε se encuentran en el cuadro nº 20.</p>

<p style="font-size: 15px;text-align: justify;">
Rugosidad relativa: $\frac{\varepsilon}{D}$</p>

<p style="font-size: 15px;text-align: justify;">
3 casos de comportamiento de una tubería:</p>

<p style="font-size: 15px;text-align: justify;">
<b>Comportamiento liso:</b> El espesor de la subcapa laminar es netamente superior a la rugosidad media del material. ($\delta_{sl} >> \varepsilon$)</p>
    
<br><br>
    
<img src="../img/comportamiento_liso.png" style="height:80px">
    
<br><br>
    
<p style="font-size: 15px;text-align: justify;">
<b>Comportamiento rugoso:</b> El espesor de la subcapa laminar es netamente inferior a la rugosidad media del material. ($\delta_{sl} << \varepsilon$)</p>
    
<br><br>
    
<img src="../img/comportamiento_rugoso.png" style="height:80px">
    
<br><br>
    
<p style="font-size: 15px;text-align: justify;">
<b>Comportamiento semirrugoso:</b> El espesor de la subcapa laminar es del mismo orden de valores de $\varepsilon$.</p>
    
<br><br>
    
<img src="../img/comportamiento_semirrugoso.png" style="height:80px">
    
<br>
    
<p style="font-size: 15px;text-align: justify;">
Para determinar el comportamiento hidráulico de una tubería nos basamos en los Reynolds frontera:</p>

<p style="font-size: 15px;text-align: justify;">
<b>Frontera lisa-semirrugosa:</b></p>

<p style="font-size: 15px;text-align: justify;">
\begin{equation}
    Re’ = \frac{23}{\frac{\varepsilon}{D}}
\end{equation}</p>

<p style="font-size: 15px;text-align: justify;">
<b>Frontera semirrugosa-rugosa:</b></p>

<p style="font-size: 15px;text-align: justify;">
\begin{equation}
    Re’’ = \frac{560}{\frac{\varepsilon}{D}}
\end{equation}</p>

<p style="font-size: 15px;text-align: justify;">
En el caso de flujo turbulento:</p>

<p style="font-size: 15px;text-align: justify;">
- Comportamiento liso: $Re < Re’$ </p>

<p style="font-size: 15px;text-align: justify;">
- Comportamiento semirrugoso: $Re’ \leq Re \leq Re’’$ </p> 

<p style="font-size: 15px;text-align: justify;">    
- Comportamiento rugoso: $Re > Re’’$ </p> 
    
<p style="font-size: 15px;text-align: justify;"> 
Para determinar el comportamiento de una tubería es recomendable utilizar el ábaco de Moody.</p>

<h1><b style="font-size: 18px;">DEMOSTRACIÓN DE DARCY-WEISBACH ∼ ANÁLISIS DIMENSIONAL PI-BUCKINGHAM</b></h1>
<br>
<hr>  
    
<p style="font-size: 15px;text-align: justify;">La pérdida de carga depende de $\rho$, $V$, $D$, $\mu$, $\varepsilon$ y $\frac{\vartriangle p}{l}$. </p>
    
<p style="font-size: 15px;text-align: justify;">
\begin{equation}
    f=\left(\frac{\vartriangle p}{l}, \rho, V, D, \mu, \varepsilon \right)
\end{equation}</p>
    
<p style="font-size: 15px;text-align: justify;">Estructura dimensional de las variables: </p>
    
<p style="font-size: 15px;text-align: justify;">
\begin{equation}
    \left[ \rho \right] = \left[ M L^{-3} T \right]\\
    \left[ V \right] = \left[ M^{0} L T^{-1} \right]\\
    \left[ D \right] = \left[ \varepsilon \right]= \left[ M^{0} L T^{0} \right]\\
    \left[ \mu \right] = \left[ M L^{-1} T^{-1} \right]\\
    \left[ \frac{\vartriangle p}{l} \right] = \left[ M^{0} L^{-2} T^{-2} \right]\\
\end{equation}</p>
    
<p style="font-size: 15px;text-align: justify;">Tomamos $\left(\rho, V, D\right)$ como variables repetidas:</p>
    
<p style="font-size: 15px;text-align: justify;">
\begin{equation}
    \left[ \pi_{_{1}} \right] = {\left[ M L^{-3} T \right]}^{a} {\left[ M^{0} L T^{-1} \right]}^{b} {\left[ M^{0} L T^{0} \right]}^{c} \left[ M L^{-1} T^{-1} \right] \rightarrow a=-1; b=-1; c=-1 \rightarrow \pi_{_{1}}= \frac{\mu}{\rho V D} \rightarrow \pi_{_{1}}= \frac{1}{Re}\\
    \left[ \pi_{_{2}} \right] = {\left[ M L^{-3} T \right]}^{a} {\left[ M^{0} L T^{-1} \right]}^{b} {\left[ M^{0} L T^{0} \right]}^{c} \left[ M^{0} L^{-2} T^{-2} \right] \rightarrow a=-1; b=-2; c=1 \rightarrow \pi_{_{2}}= \frac{\frac{\vartriangle p}{l} \cdot D}{\rho V^{^{2}}}\\
    \left[ \pi_{_{3}} \right] = {\left[ M L^{-3} T \right]}^{a} {\left[ M^{0} L T^{-1} \right]}^{b} {\left[ M^{0} L T^{0} \right]}^{c} \left[ M^{0} L T^{0} \right] \rightarrow \pi_{_{3}} = \frac{\varepsilon}{D}= \varepsilon_{_{r}} \hspace{2mm} (rugosidad \hspace{1mm} relativa)\\
\end{equation}</p>
    
<br>
    
<p style="font-size: 15px;text-align: justify;">
\begin{equation}
    \phi \left(\frac{1}{Re},\frac{\frac{\vartriangle p}{l} \cdot D}{\rho V^{^{2}}},\varepsilon_{_{r}}\right)=0 \rightarrow \frac{\frac{\vartriangle p}{l} \cdot D}{\rho V^{^{2}}} = f \left( \frac{1}{Re},\varepsilon_{_{r}}\right) \rightarrow {\vartriangle{p}} = \frac{l \cdot \rho \cdot V^{^{2}}}{D} f \left( \frac{1}{Re},\varepsilon_{_{r}}\right)
\end{equation}</p>    
    
<p style="font-size: 15px;text-align: justify;">
\begin{equation}
    h_{_{f}}= \frac{\vartriangle p}{\rho \cdot g} = \frac{l \cdot \rho \cdot V^{^{2}}}{2 \cdot D \cdot g} 2 \cdot f \left( \frac{1}{Re},\varepsilon_{_{r}}\right) \rightarrow 2 \cdot f \left( \frac{1}{Re},\varepsilon_{_{r}}\right) = f \rightarrow \boxed{h_{_{f}}= f \cdot \frac{L}{D} \cdot \frac{V^{^{2}}}{2 \cdot g}}
\end{equation}</p>
